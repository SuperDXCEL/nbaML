# Import the DataFrame with all of LeBron's games played in the last 10 seasons

```python
# all_games_df = pd.read_csv('/kaggle/input/final-statistics-97000-x-157/allplayerstatistics.csv')
all_games_df = pd.read_csv('/kaggle/input/almost-last-10-seasons/almostLastTenSeasons.csv')
all_games_df
```

# Drop unnecessary columns

```python
cols_to_drop = ['WL', 'GAME_DATE', 'Game_ID']
all_games_df.drop(columns=cols_to_drop, inplace=True)
```

# Divide categorical and continuous columns

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

LE = LabelEncoder()
SS = StandardScaler()
categorical = ['TEAM', 'RIVAL', 'TEAM_NAME', 'RIVAL_NAME']
continuous = ['MIN', 'FGM', 'FGA', 'FG_PCT', 'FG3M', 'FG3A', 'FG3_PCT', 'FTM', 'FTA',
       'FT_PCT', 'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 'TOV', 'PF',
       'PTS', 'PLUS_MINUS']

for col in categorical:
    all_games_df[col] = LE.fit_transform(all_games_df[col])
for col in continuous:
    all_games_df[col] = SS.fit_transform(all_games_df[col].values.reshape(-1, 1))
```

```python
all_games_df.isnull().sum()
```

# Take care of missing values

```python
# Handle rank columns
rank_columns = [col for col in all_games_df.columns if 'RANK' in col]

for col in rank_columns:
    all_games_df[col].fillna(all_games_df[col].max() + 1, inplace=True)

for col in rank_columns:
    unique_ranks = sorted(all_games_df[col].unique(), reverse=True)
    all_games_df[col] = pd.Categorical(all_games_df[col], categories=unique_ranks, ordered=True)
```

```python
pd.options.display.max_rows = None
all_games_df.isnull().sum()
```

```python
# Handle the final three with missing values
cols_with_null = ['TEAM_W_PCT', 'RIVAL_TEAM_ID', 'RIVAL_W_PCT', 'RIVAL_SEASON']

# TEAM_W_PCT
winning_mode = all_games_df['TEAM_W_PCT'].mode()[0]
all_games_df['TEAM_W_PCT'].fillna(winning_mode, inplace=True)

# TEAM_SEASON
all_games_df['TEAM_SEASON'].fillna(-1, inplace=True)
# RIVAL_SEASON
all_games_df['RIVAL_SEASON'] = all_games_df['TEAM_SEASON']

# RIVAL_TEAM_ID
all_games_df['RIVAL_TEAM_ID'].fillna(-1, inplace=True)
                  

# RIVAL_W_PCT
rival_winning_mode = all_games_df['RIVAL_W_PCT'].mode()[0]
all_games_df['RIVAL_W_PCT'].fillna(rival_winning_mode, inplace=True)
```

# Prepare the training and validation datasets

```python
from sklearn.model_selection import train_test_split

y = all_games_df[continuous].copy()
X = all_games_df.drop(columns=list(y.columns))

# Apply pd.get_dummies to the selected columns
dummyColumns = ['TEAM_SEASON', 'RIVAL_SEASON']
encoded_df = pd.get_dummies(all_games_df[dummyColumns], drop_first=True)

# Concatenate the original DataFrame with the encoded columns
X = pd.concat([X.drop(columns=dummyColumns), encoded_df], axis=1)

X_train, X_valid, y_train, y_valid = train_test_split(X, y, random_state=42)
```
